```python
import pandas as pd
from pathlib import Path

#Load File

current_data = Path('Data/alt_fuel_stations.csv')

#Convert data to dataframe with Pandas

current_df = pd.read_csv(current_data)

current_df.head()
```

```python
#Remove unwanted columns

to_drop = ['Station Name', 'Street Address', 'Intersection Directions', 'City', 'ZIP', 'Plus4', 
           'Station Phone', 'Status Code', 'Restricted Access', 'RD Blends', 'RD Blends (French)', 
           'RD Blended with Biodiesel', 'RD Maximum Biodiesel Level', 'NPS Unit Name', 
           'CNG Station Sells Renewable Natural Gas', 'LNG Station Sells Renewable Natural Gas', 
           'Maximum Vehicle Class', 'EV Workplace Charging', 'Expected Date', 'Groups With Access Code', 
           'Access Days Time', 'Cards Accepted', 'BD Blends', 'NG Fill Type Code', 'NG PSI', 'EV Pricing', 
           'EV Pricing (French)', 'LPG Nozzle Types', 'Hydrogen Pressures', 'Hydrogen Standards', 'CNG Fill Type Code', 
           'CNG PSI', 'CNG Vehicle Class', 'LNG Vehicle Class', 'EV On-Site Renewable Source', 'EV Other Info', 
           'Geocode Status', 'Access Code', 'Access Detail Code', 'Federal Agency Code', 'Facility Type', 
           'CNG Total Compression Capacity', 'CNG Storage Capacity', 'LNG On-Site Renewable Source', 'EV Network', 
           'Country', 'Date Last Confirmed', 'EV Connector Types', 'Intersection Directions (French)', 
           'Access Days Time (French)', 'BD Blends (French)', 'Groups With Access Code (French)', 'EV Level1 EVSE Num', 
           'EV Level2 EVSE Num', 'EV DC Fast Count', 'EV Network Web', 'ID', 'Updated At', 'Federal Agency ID', 
           'Federal Agency Name', 'Open Date', 'Hydrogen Status Link', 'NG Vehicle Class', 'LPG Primary', 'E85 Blender Pump', 
           'Hydrogen Is Retail', 'CNG On-Site Renewable Source', 'E85 Other Ethanol Blends', 'Owner Type Code', 'Latitude', 
           'Longitude', 'CNG Dispenser Num']

current_df = current_df.drop(to_drop, axis = 1)
```

```python
#Confirm Column Drop

current_df.head()
```

```python
#Group fuel types and assign by state

totals_df = current_df.groupby('State')['Fuel Type Code'].value_counts().unstack(fill_value = 0)


totals_df.head()            
```

```python
#Export new data frame

totals_df.to_csv('Current_Alt_Fuel_stations.csv')
```
